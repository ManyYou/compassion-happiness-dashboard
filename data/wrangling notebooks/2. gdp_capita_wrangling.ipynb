#### Check the Structure

```python
# Remove leading & trailing whitespace + replace multiple spaces within names to a single underscore
df_gdp.columns = df_gdp.columns.str.strip().str.replace(r'\s+', '_', regex=True)

# Duplicate
# Automatically detect categorical/object columns
cat_cols = df_gdp.select_dtypes(include=['object', 'category']).columns
# Strip leading and trailing spaces from all categorical columns
df_gdp[cat_cols] = df_gdp[cat_cols].apply(lambda x: x.str.strip())

# Check dataset structures
print("Dataset Structures: ")
display(df_gdp.info())

# First 5 rows
print("\nFirst 5 rows: ")
display(df_gdp.head())

# Last 5 rows
print("\nLast 5 rows: ")
display(df_gdp.tail())

# Check the descriptive statistics of the datasete
print('\nCategorical Variables:')
display(df_gdp.describe(include = 'object'))

print('\nNumerical Variables: ')
display(df_gdp.describe(include = 'number'))

# Check missing values 
print("\nMissing values count for each column: ")
display(df_gdp.isnull().sum())  

print("\nMissing values percentage: ")
display(df_gdp.isna().mean() * 100)   

num_duplicates = df_gdp.duplicated().sum()
print(f"\nTotal duplicate rows: {num_duplicates}")
```

#### Select Necessary Columns

```python
# Select only these columns (subset)
selected_cols = ['Country_Name', 'Country_Code', '2018', '2019', '2020', '2021', '2022']

df_gdp = df_gdp[selected_cols]
```

#### Renaming Columns

```python
df_gdp.rename(columns={
    'Country_Name': 'country',
    'Country_Code': 'country_code'
}, inplace=True)
```

#### Convert Wide to Long

```python
years = ['2018', '2019', '2020', '2021', '2022']

df_gdp = df_gdp.melt(
    id_vars=['country', 'country_code'],   # Include Country Code
    value_vars= years,                          # Columns to unpivot
    var_name='year_gdp_capita',                           # New column for years
    value_name='gdp_per_capita'                # New column for values
)

df_gdp = df_gdp.dropna(subset=['gdp_per_capita'])
df_gdp['year_gdp_capita'] = df_gdp['year_gdp_capita'].astype(int)

df_gdp = df_gdp.reset_index(drop=True)
```

#### Handle Multiple Years

```python
# Filter only rows with year ≤ 2022
df_gdp = df_gdp[df_gdp['year_gdp_capita'] <= 2022]

# For each country, get its latest year
latest_year_per_country = df_gdp.groupby('country')['year_gdp_capita'].max().reset_index()

# Merge back to get full rows matching latest year for each country
df_gdp = df_gdp.merge(latest_year_per_country, on=['country', 'year_gdp_capita'], how='inner')

# Drop year
df_gdp.drop(columns=["year_gdp_capita"], inplace=True)
```
