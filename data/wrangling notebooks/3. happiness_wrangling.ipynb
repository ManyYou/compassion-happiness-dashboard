```python
# Necessary Libraries
import pandas as pd
import re
```

#### Load the Dataset


```python
df_happiness = pd.read_excel('world_happiness_report_2022.xls')
```

```python
df_happiness.shape
```

```python
# Remove leading & trailing whitespace + replace multiple spaces within names to a single underscore
df_happiness.columns = df_happiness.columns.str.strip().str.replace(r'\s+', '_', regex=True)

# Duplicate
# Automatically detect categorical/object columns
cat_cols = df_happiness.select_dtypes(include=['object', 'category']).columns
# Strip leading and trailing spaces from all categorical columns
df_happiness[cat_cols] = df_happiness[cat_cols].apply(lambda x: x.str.strip())

# Check dataset structures
print("Dataset Structures: ")
display(df_happiness.info())

# First 5 rows
print("\nFirst 5 rows: ")
display(df_happiness.head())

# Last 5 rows
print("\nLast 5 rows: ")
display(df_happiness.tail())

# Check the descriptive statistics of the datasete
print('\nCategorical Variables:')
display(df_happiness.describe(include = 'object'))

print('\nNumerical Variables: ')
display(df_happiness.describe(include = 'number'))

# Check missing values 
print("\nMissing values count for each column: ")
display(df_happiness.isnull().sum())  

print("\nMissing values percentage: ")
display(df_happiness.isna().mean() * 100)   

num_duplicates = df_happiness.duplicated().sum()
print(f"\nTotal duplicate rows: {num_duplicates}")
```

#### Drop Last Row (NULL)

```python
df_happiness = df_happiness.drop(index=146)  # Drop row with index 3
```

#### Select a Subset of Column

```python
# Select only these columns (subset)
selected_cols = ['Country', 'Happiness_score']

df_happiness = df_happiness[selected_cols]
```

#### Rename Column

```python
df_happiness.rename(columns={
    'Country': 'country',
    'Happiness_score': 'happiness_score'
}, inplace=True)
```

```python
df_happiness.info()
```

#### Clean Country Names

```python
df_happiness['country'] = df_happiness['country'].str.replace('*', '', regex=False).str.strip()
```

#### Merge with the other datasets

```python
df_gdp_gini = pd.read_csv('gini_gdp.csv')
```

```python
df_gdp_gini['country'].unique()
```

```python
df_happiness['country'].unique()
```

```python
# Get unique values from both datasets
set1 = set(df_happiness['country'].unique())
set2 = set(df_gdp_gini['country'].unique())

# Values in df1 but not in df2
diff_1_not_in_2 = set1 - set2

# Values in df2 but not in df1
diff_2_not_in_1 = set2 - set1

print("Country in df_happiness but not in df_gdp_gini: \n", diff_1_not_in_2)
print("\nCountry in df_gdp_gini but not in df_happiness: \n", diff_2_not_in_1)
```

```python
country_mapping = {
    'Palestinian Territories': 'Palestine',
    "Cote d'Ivoire": 'Ivory Coast',
    'North Cyprus': 'Cyprus',
    'Hong Kong S.A.R. of China': 'Hong Kong',
    'Taiwan Province of China': 'Taiwan',
    'Eswatini, Kingdom of': 'Eswatini',
    'Kosovo': 'Kosovo',  # only if named differently elsewhere
}

df_happiness['country'] = df_happiness['country'].replace(country_mapping)
df_gdp_gini['country'] = df_gdp_gini['country'].replace(country_mapping)
```

```python
df_merged = pd.merge(
    df_gdp_gini,
    df_happiness,
    on='country',  # Common column for joining
    how='inner'         # Inner Join: keep only matching rows
)
```
