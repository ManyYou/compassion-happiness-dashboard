```python
# Remove leading & trailing whitespace + replace multiple spaces within names to a single underscore
df_oecd.columns = df_oecd.columns.str.strip().str.replace(r'\s+', '_', regex=True)

# Duplicate
# Automatically detect categorical/object columns
cat_cols = df_oecd.select_dtypes(include=['object', 'category']).columns
# Strip leading and trailing spaces from all categorical columns
df_oecd[cat_cols] = df_oecd[cat_cols].apply(lambda x: x.str.strip())

# Check dataset structures
print("Dataset Structures: ")
display(df_oecd.info())

# First 5 rows
print("\nFirst 5 rows: ")
display(df_oecd.head())

# Last 5 rows
print("\nLast 5 rows: ")
display(df_oecd.tail())

# Check the descriptive statistics of the datasete
print('\nCategorical Variables:')
display(df_oecd.describe(include = 'object'))

print('\nNumerical Variables: ')
display(df_oecd.describe(include = 'number'))

# Check missing values 
print("\nMissing values count for each column: ")
display(df_oecd.isnull().sum())  

print("\nMissing values percentage: ")
display(df_oecd.isna().mean() * 100)   

num_duplicates = df_oecd.duplicated().sum()
print(f"\nTotal duplicate rows: {num_duplicates}")
```

#### Select a Subset of Columns

```python
# Select only these columns (subset)
selected_cols = ['REF_AREA', 'Reference_area', 'TIME_PERIOD', 'OBS_VALUE']

df_oecd = df_oecd[selected_cols]
```

#### Rename Columns

```python
df_oecd.rename(columns={
    'REF_AREA': 'country_code',
    'Reference_area': 'country',
    'TIME_PERIOD': 'year_oecd',
    'OBS_VALUE': 'percent_gdp_social_spending'
}, inplace=True)
```

#### Get the latest year 2022 or most recent before

```python
# Filter only rows with year ≤ 2022
df_oecd = df_oecd[df_oecd['year_oecd'] <= 2022]

# For each country, get its latest year
latest_year_per_country = df_oecd.groupby('country')['year_oecd'].max().reset_index()

# Merge back to get full rows matching latest year for each country
df_oecd = df_oecd.merge(latest_year_per_country, on=['country', 'year_oecd'], how='inner')

# Drop Year
df_oecd.drop(columns=["year_oecd"], inplace=True)
```
